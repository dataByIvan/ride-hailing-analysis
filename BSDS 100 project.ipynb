### 🧾 Step 1: Understand the Columns (Simple Explanations)

Here’s what the main columns in the **NYC Yellow Taxi dataset** mean in plain English:

| Column | Meaning |
|:--|:--|
| **VendorID** | The taxi company that provided the ride (usually 1 or 2). |
| **tpep_pickup_datetime** | The date and time when the passenger was picked up. |
| **tpep_dropoff_datetime** | The date and time when the passenger was dropped off. |
| **passenger_count** | Number of passengers in the taxi. |
| **trip_distance** | How far the trip was, in miles. |
| **RatecodeID** | Type of rate charged (1 = Standard, 2 = JFK, 3 = Newark, etc.). |
| **store_and_fwd_flag** | Whether the trip record was sent to the server later (N = No, Y = Yes). |
| **PULocationID** | Pickup location zone ID (zones are numbered across NYC). |
| **DOLocationID** | Drop-off location zone ID. |
| **payment_type** | How the passenger paid (1 = Credit card, 2 = Cash, 3 = No charge, etc.). |
| **fare_amount** | Base fare before extras, tolls, or tips. |
| **extra** | Additional fees (like night surcharge or peak-hour fee). |
| **mta_tax** | NYC tax on the ride. |
| **tip_amount** | Tip amount (0 if paid in cash or not tipped). |
| **tolls_amount** | Total tolls paid during the trip. |
| **improvement_surcharge** | Standard $0.30 fee added to every trip. |
| **total_amount** | Total fare (fare + tips + tolls + taxes + surcharges). |
| **congestion_surcharge** | Extra fee applied in traffic congestion zones (recent years). |
